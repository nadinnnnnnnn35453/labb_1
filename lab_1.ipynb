```python
def task_4():
    try:
        input_str = input("Введіть список чисел через пробіл: ")
        numbers = list(map(int, input_str.split()))
        
        swap_count = 0
        n = len(numbers)
        for i in range(n):
            for j in range(0, n-i-1):
                if numbers[j] > numbers[j+1]:
                    # Якщо поточний елемент більший за наступний, міняємо їх місцями
                    numbers[j], numbers[j+1] = numbers[j+1], numbers[j]
                    swap_count += 1
                    
        print(f"Було виконано {swap_count} переставлень")
    
    except ValueError:
        print("Помилка. У списку присутнє нечислове значення.")
task_4()
```
